```python
import xml.etree.ElementTree as ET
import os
from inspect import getmembers, isfunction, isclass
```

```python
ead_file = os.path.join(r'C:\Users\madis\Documents\676nlam\si676-2024-data\data\xml\day_20221004_205435_UTC__ead.xml')
```

```python
tree = ET.parse(ead_file)
root = tree.getroot()

for element in root:
    print(element.tag, type(element))
```

```python
print(root[:150])
```

```python
for element in root.iter():
    print(element.tag, element.attrib)
```

```python
archDesc = root.find('{http://ead3.archivists.org/schema/}archdesc')
```

```python
if archDesc is not None:
    for element in archDesc:
        print(element.tag, element.attrib)
else:
    print("archDesc element not found")
```
